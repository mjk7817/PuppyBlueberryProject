```python
path = Path('drive/My Drive/archive')
```

```python
fnames = get_image_files(path)
```

```python
fnames
```

```python
np.random.seed(2)
pat = re.compile(r'[/\\]([^/\\]+)\d+.jpg$') #find expression in picture title
bs=10#test a batch size (cannot be 16 due to size constraint)
```

```python
# fastai data bunch 
data=ImageDataBunch.from_name_re(path,fnames,pat,ds_tfms=get_transforms(),size=224,bs=bs).normalize(imagenet_stats)
```

```python
data.show_batch(rows=5,figsize=(10,10))
```
